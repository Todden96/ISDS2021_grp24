Functions intended for descriptive characteristics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def TagsViolinplotted(df, feature_list, def_value=1, price="price"):
    passed = []
    for f in feature_list:
        if df[df[f]==def_value]["price"].mean() > 0:
            passed.append(f)
    
    fig, axs = plt.subplots(1, len(passed), figsize=(18, 10), sharey=True, sharex=True)
    fig_nr = 0
    
    for p in passed: 
        sns.violinplot(y=df[df[p]==def_value][price], data = df[df[p]==def_value], ax=axs[fig_nr])
        #axs[fig_nr] = sns.despine()
        axs[fig_nr].set(xlabel=p)
        axs[fig_nr].set(ylabel="")
        fig_nr += 1
    axs[0].set(ylabel="Price in US Dollars")
    #plt.figure(figsize=(20, 3))  # width:20, height:3
    #plt.bar(range(len(passed)*2), df[price], align='edge', width=0.5)
    #plt.bar(range(len(passed)), df[price], align="center")
    #plt.xticks(range(len(passed)))
    plt.show
```

Finding the top 10 tags

```python
all_tags = list(df.columns[35:-4])

tags, counts = [], []

for t in all_tags:
    count = df[t].mean()
    counts.append(count)
    tags.append(t)

tdict = (dict(zip(tags, counts)))
```

```python
from operator import itemgetter

top10 = dict(sorted(tdict.items(), key = itemgetter(1), reverse = True)[:10])
top10
```

```python
TagsViolinplotted(df, top10, def_value=1, price="price")
#plt.xticks(rotation=70)
```
